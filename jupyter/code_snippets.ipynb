```python
import json
import math
import pandas as pd
```

```python
# loading data into a Pandas DataFrame
# NB: this is one of MANY ways Pandas provides
filepaths = ['../data/dns.json']
df = pd.concat([pd.read_json(file, lines=True) for file in filepaths], ignore_index=True)    
df.head()
```

```python
# discovering column names 
df.columns.values
```

```python
# calculating a new column from an existing column
df['new_col'] = df['rtt']*2
df['new_col2'] = df['new_col'].apply(lambda x: math.ceil(x) if not pd.isna(x) else x)
#renaming columns
df = df.rename(columns={'new_col':'dbl_ttl', 'new_col2':'ceil_dbl_ttl'})
# selecting a subset of columns
df[['dbl_ttl', 'ceil_dbl_ttl']].head()
```

```python
# grouping and sorting in a Pandas DataFrame
grouped = df.groupby('id_orig_h').sum().reset_index()
sorted = grouped.sort_values('ceil_dbl_ttl', ascending=False)
sorted[['id_orig_h', 'rtt']].head()
```
